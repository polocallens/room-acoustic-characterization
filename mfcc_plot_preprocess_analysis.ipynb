```python
from mfcc import *
from librosa import display
import matplotlib.pyplot as plt
```

```python
%cd ../RT-estimation/RCNN/Datasets/Music/
```

```python
m_sr, music = wavfile.read('smaller2_fma_all_wav_sox_normed_8s/000540.wav')
```

```python
def compute_norm_mfcc_3(signal,fs):
    signal = feature_normalize(signal)
    mfcc = librosa.feature.mfcc(y=signal, sr=fs, n_mfcc=40, )
    #mfcc = librosa.util.fix_length(mfcc, window_size, axis=1, mode='wrap') #Reshape to windom size
    mfcc = librosa.util.normalize(mfcc, axis=1)
    return mfcc
```

```python
mfcc = compute_norm_mfcc_3(music,m_sr)
```

```python

librosa.display.specshow(mfcc)
plt.colorbar(format='%+2.0f dB')
plt.title('Log-frequency power spectrogram')
```
